# Here we test the time independent Raman Hamiltonian

```python
def unify(st):
    return st/st.norm()

def ketbra(ket,bra=None):
    if not bra:
        bra = ket
    return ket * bra.dag()

L1 = unify(ket('110')+ket('000'))
L2 = unify(ket('111')-ket('001'))
L3 = unify(ket('110')-ket('000'))
I = tensor(qeye(2),qeye(2),qeye(2))

#Error Correction
err = [tensor(sigmax(),qeye(2),qeye(2)),tensor(sigmay(),qeye(2),qeye(2))]
Pe = sum([ketbra(op*L1) + ketbra(op*L2) for op in err])
U = sum([ketbra(op*L1,L1) + ketbra(op*L2,L2) for op in err])
U = U + U.dag()
Pc = I - Pe

class params(object):
    def __init__(self):
        # Parameters
        self.g = 0.1
        self.gamma = 0.1
        self.gec = 0.1
        self.delta = 10
    def rho0(self):
        return ketbra(L1)
    def Ht(self):
        #Hamiltonian and changes
        H0 = -self.delta*(ketbra(L1)+ketbra(L2))
        V = self.g/2 * (ketbra(L1,L3) + ketbra(L2,L3))
        V = V + V.dag()
        return H0 + V
    def Ld(self):
        ld = lindblad_dissipator(tensor(sigmam(),qeye(2),qeye(2)))
        # Add upwords dissipation
        ld += lindblad_dissipator(tensor(sigmap(),qeye(2),qeye(2)))
        return self.gamma * ld
    def Lec(self):
        lec = sprepost(Pc,Pc)+ sprepost(U*Pe,Pe*U) -spre(I)
        return  self.gec * lec

class rhoResult(object):
    def __init__(self,data,rho,timelist=None):
        if timelist==None:
            timelist = xrange(len(rho))
        self.data = data
        self.rho = rho
        self.timelist = timelist
    def oo(self):
        o1 = expect(ketbra(L1),self.rho)
        o2 = expect(ketbra(L2),self.rho)
        return [o1,o2]
    def pop(self):
        [o1,o2] = self.oo()
        plot(self.timelist, real(o1), 'b',self.timelist, real(o2), 'r')
        xlabel('Time')
        ylabel('Occupation probability')
        title('Occupation of L1 and L2 vs time')        
    def decay(self):
        eigs = [real(b.eigenenergies()[-1]) for b in self.rho]
        xlabel('Time')
        ylabel('biggest eigenvalue')
        title('decay rate')
        plot(self.timelist,eigs)

def exper(data=params(),eflag=True,cflag=True,time=None,steps=1000):
    c_op_list = []
    if eflag:
        c_op_list.append(data.Ld())
    if cflag:
        c_op_list.append(data.Lec())
    if not time:
        time = 10*2*pi/data.g
    timelist = linspace(0,time,steps)
    out = mesolve(data.Ht(),data.rho0(),timelist,c_op_list,[])
    return rhoResult(data,out.states,timelist)
```

```python
d = params()
d.gamma = 0
a = exper(cflag=False,eflag=False,data=d,time=10*2*pi/(d.g**2/d.delta),steps=10000)
a.pop()
```

```python
d = params()
d.gamma = 0
d.delta = 1
a = exper(cflag=False,eflag=False,data=d,time=10*2*pi/(d.g**2/d.delta),steps=10000)
a.pop()
```

# Trying with $\sigma_{z}$ instead of $\left|L_{1}\right\rangle\left\langle L_{3}\right|$ and $\sigma_{x}$ instead of $\left|L_{2}\right\rangle\left\langle L_{3}\right|$

```python
class paramz(params):
    def Ht(self):
        #Hamiltonian and changes
        L4 = unify(ket('111') + ket('001'))
        H0 = -self.delta*(ketbra(L1)+ketbra(L2)) + self.delta*10*ketbra(L4)
        sz = tensor(sigmaz(),qeye(2),qeye(2))
        sx3 = tensor(qeye(2),qeye(2),sigmax())
        V = self.g/4*(sz+sx3)
        V = V + V.dag()
        return H0 + V
d = paramz()
d.delta = 1
d.gec = 10*d.gamma
a = exper(cflag=True,eflag=True,data=d,time=10*2*pi/(d.g**2/d.delta),steps=10000)
a.pop()
```
