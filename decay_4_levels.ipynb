```python
def ketbra(i,j):
    return ket('%d'%i,4)*bra('%d'%j,4)

def operate(a,b):
    return vector_to_operator(a*operator_to_vector(b))
```

```python
class params:
    def __init__(self):
        # Parameters
        self.g = 1
        self.gamma = 0.1
        self.gec = 0.1
    def Ht(self):
        #Hamiltonian and changes
        H0 = self.g * (ketbra(0,1)+ketbra(1,0))
        return H0
    def Ld(self):
        ld = lindblad_dissipator(ketbra(2,0))+lindblad_dissipator(ketbra(3,1))
        return self.gamma * ld
    def Lec(self):
        Pc = ketbra(0,0) + ketbra(1,1)
        Pe = ketbra(2,2) + ketbra(3,3)
        U = ketbra(0,2) + ketbra(1,3)
        U = U + U.dag()
        lec = sprepost(Pc,Pc)+ sprepost(U*Pe,Pe*U) -spre(qeye(4))
        return  self.gec * lec
```

```python
def chunks(l, n):
    """ Yield successive n-sized chunks from l.
    """
    for i in xrange(0, len(l), n):
        yield l[i:i+n]
```

```python
import copy
class rhoResult:
    def __init__(self,data,rho,timelist=None):
        if timelist==None:
            timelist = xrange(len(rho))
        self.data = copy.copy(data)
        self.rho = rho
        self.timelist = timelist
    def oo(self):
        o1 = expect(ketbra(0,0),self.rho)
        o2 = expect(ketbra(1,1),self.rho)
        o3 = expect(ketbra(2,2),self.rho)
        o4 = expect(ketbra(3,3),self.rho)
        return [o1,o2,o3,o4]
    def pop(self):
        [o1,o2,o3,o4] = self.oo()
        plot(self.timelist, real(o1), 'b',self.timelist, real(o2), 'r',
             self.timelist,real(o3),'g',self.timelist,real(o4),'y')
        xlabel('Time')
        ylabel('Occupation probability')
        title('Occupation of L1 and L2 and Le vs time')        
    def decay(self):
        eigs = [real(b.eigenenergies()[-1]) for b in self.rho]
        xlabel('Time')
        ylabel('biggest eigenvalue')
        title('decay rate')
        plot(self.timelist,eigs)
```

```python
def exper(data=params(),eflag=True,cflag=True,steps=1000):
    c_op_list = []
    if eflag:
        c_op_list.append(data.Ld())
    if cflag:
        c_op_list.append(data.Lec())
    timelist = linspace(0,10*2*pi/data.g,steps)
    rho0 = ketbra(0,0)
    out = mesolve(data.Ht(),rho0,timelist,c_op_list,[])
    return rhoResult(data,out.states,timelist)
```

```python
def unify(st):
    return st/st.norm()

def ketbra(ket,bra=None):
    if not bra:
        bra = ket
    return ket * bra.dag()

L1 = unify(ket('110')+ket('000'))
L2 = unify(ket('111')-ket('001'))
I = tensor(qeye(2),qeye(2),qeye(2))

#Error Correction
err = [tensor(sigmax(),qeye(2),qeye(2)),tensor(sigmay(),qeye(2),qeye(2))]
Pe = sum([ketbra(op*L1) + ketbra(op*L2) for op in err])
U = sum([ketbra(op*L1,L1) + ketbra(op*L2,L2) for op in err])
U = U + U.dag()
Pc = I - Pe

class params:
    def __init__(self):
        # Parameters
        self.g = 0.1
        self.gamma = 0.1
        self.gec = 0.1
    def rho0(self):
        return ketbra(L1)
    def Ht(self):
        #Hamiltonian and changes
        H0 = self.g * (L1*L2.dag() + L2*L1.dag())
        return H0
    def Ld(self):
        ld = lindblad_dissipator(tensor(sigmam(),qeye(2),qeye(2)))
        # Add upwords dissipation
        ld += lindblad_dissipator(tensor(sigmap(),qeye(2),qeye(2)))
        return self.gamma * ld
    def Lec(self):
        lec = sprepost(Pc,Pc)+ sprepost(U*Pe,Pe*U) -spre(I)
        return  self.gec * lec

class rhoResult:
    def __init__(self,data,rho,timelist=None):
        if timelist==None:
            timelist = xrange(len(rho))
        self.data = data
        self.rho = rho
        self.timelist = timelist
    def oo(self):
        o1 = expect(ketbra(L1),self.rho)
        o2 = expect(ketbra(L2),self.rho)
#        o3 = expect(ketbra(2,2),self.rho)
#        o4 = expect(ketbra(3,3),self.rho)
        return [o1,o2]#,o3,o4]
    def pop(self):
        [o1,o2] = self.oo()
        plot(self.timelist, real(o1), 'b',self.timelist, real(o2), 'r')#,
#             self.timelist,real(o3),'g',self.timelist,real(o4),'y')
        xlabel('Time')
        ylabel('Occupation probability')
        title('Occupation of L1 and L2 vs time')        
    def decay(self):
        eigs = [real(b.eigenenergies()[-1]) for b in self.rho]
        xlabel('Time')
        ylabel('biggest eigenvalue')
        title('decay rate')
        plot(self.timelist,eigs)

def exper(data=params(),eflag=True,cflag=True,time=None,steps=1000):
    c_op_list = []
    if eflag:
        c_op_list.append(data.Ld())
    if cflag:
        c_op_list.append(data.Lec())
    if not time:
        time = 10*2*pi/data.g
    timelist = linspace(0,time,steps)
    out = mesolve(data.Ht(),data.rho0(),timelist,c_op_list,[])
    return rhoResult(data,out.states,timelist)
```
