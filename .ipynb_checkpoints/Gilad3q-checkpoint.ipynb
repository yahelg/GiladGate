# Simulating Gilad-Gate with three qubits only. The far qubit is ommitted.

## Copying the conclusing from DecayAndCorrect:

<h3> Conclusions so far:
- The population of $\left|L_3\right>$ is $\frac{1}{2}\left(\frac{g}{\Delta}\right)^2$
- The decay rate is $\frac{1}{2}\gamma\left(\frac{g}{\Delta}\right)^2$
- This means the number of oscillations ~ $\frac{\Delta}{\gamma}$
<br><br>
The reason why the population of  $\left|L_3\right>$ isn't affected by $\gamma$ is that the Rabi oscillations between L1 and L3 go like $\frac{g^{2}}{g^{2}+\Delta^{2}}\sin^{2}\left(\frac{\sqrt{g^{2}+\Delta^{2}}}{2}t\right)\sim\frac{g^{2}}{\Delta^{2}}\sin^{2}\left(\frac{\Delta}{2}t\right)$ which is a faster pace than $\gamma$.

## Extra conclusions:

<h3> 
- The $\frac{1}{2}\gamma\left(\frac{g}{\Delta}\right)^2$ decay is caused by the amplitude of $\left|L_3\right>$ being added after correction to that of $\left|L_1\right>$.


- Another source of decay is the double error, which has an effective decay rate of $\gamma_{eff}=\gamma\frac{\gamma}{\Gamma_{ec}+\gamma}$


- To get oscillations we need to have a higher driving rate than both decay rates:<br>

    + $\frac{g^2}{\Delta}\gg\frac{1}{2}\gamma\left(\frac{g}{\Delta}\right)^2$

    + $\frac{g^2}{\Delta}\gg\gamma\frac{\gamma}{\Gamma_{ec}+\gamma}$<br>

- If both conditions apply, we should see $~\frac{\Delta}{\gamma}$ oscillations.


- For some reason that is as yet unclear to me, the last condition can be relaxed if the state $\left|111+001\right>$ has the same energy as L3, and the driving is |L1><L3| + |L2><L3| + h.c. However if the driving is $\sigma_z$ and $\sigma_x$ this would ruin the Raman transitions.


- Right now it is left to discover the reason, and to check the validity of the conditions above.

## Conclusions about no-driving or pre-driving:

<h3> 
- Before starting the process, we get a fast decay of the time order of $1/\gamma$, to a mixed state implied by the rate equations:

$\begin{eqnarray*}
\dot{P_{1}} & = & -2\gamma P_{1}+\Gamma P_{e}+\gamma P_{e}\\
\dot{P_{3}} & = & -2\gamma P_{3}+\gamma P_{e}\\
\dot{P_{e}} & = & 2\gamma\left(P_{1}+P_{3}\right)-\left(\Gamma+2\gamma\right)P_{e}
\end{eqnarray*}$

- For $\Gamma=0$ this implies 25% population in L1, 25% in L3 and 50% in the error states.
- Otherwise we get $2P_3=P_e=\frac{2\gamma}{\gamma+\Gamma}P_1$ (got it wrong by a factor of 2 somewhere)

```python
class paramsAllOut(paramsSigma):
    def H0(self):
        a = 5*self.delta*(I-ketbra(ket('110'))
                               -ketbra(ket('000'))-ketbra(ket('111'))
                               -ketbra(ket('001')))
        a += -self.delta*(ketbra(L1)+ketbra(L2)) + 5*self.delta*ketbra((ket('111')+ket('001')).unit())
        return a
        
d = paramsAllOut()
d.g = 0.1
d.gamma = 0.1
d.delta = 1
d.gec = 10

out[len(In)-1] = exper(data=d)
a = out[len(In)-1]
dc = showdecay(a)
a.pop()
show()
pr = showperiod(a)
```
